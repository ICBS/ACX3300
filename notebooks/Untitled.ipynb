#### Data dictionary

| **Fieldnames** | **Definition**                                                 | **Description**                                                                                                                                                                                        |
|----------------|----------------------------------------------------------------|--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------|
| GVKEY      | A Unique Global Company Key                                    | A unique identifier that represents each company throughout Xpressfeed. All company data records are identified by a GVKEY. Except in rare circumstances |
| DATADATE   | Data Date                                                      | The reporting date for a data record                                                                            |
| AT             | Assets – Total/Liabilities and Stockholders’ Equity – Total    | Total Assets (Gross)                                                                                                                                                                                   |
| CEQ            | Common Equity – Total                                          | Total Book Value minus Treasury Stock – Dollar Amount – Preferred                                                                                                                                      |
| CHE            | Cash and Short-Term Investments                                | Cash Special Deposits, Working Funds and Temporary Cash Investments                                                                                                                                    |
| CSHO           | Common Shares Outstanding                                      | Number of Shares Outstanding                                                                                                                                                                           |
| DLC            | Debt in Current Liabilities                                    | Debt – Long-Term Due Within One Year plus Debt – Short-Term – Total                                                                                                                                    |
| DLTT           | Long-Term Debt – Total                                         | Debt – Long-Term – Total                                                                                                                                                                               |
| INVT           | Inventories – Total                                            | Inventories plus Gas Stored Underground                                                                                                                                                                |
| IVST           | Short-Term Investments                                         | Special Deposits, Working Funds, and Temporary Cash Investments                                                                                                                                        |
| MIB            | Redeemable Noncontrolling Interest (Balance Sheet)             | Redeemable Noncontrolling Interest (Balance Sheet)                                                                                                                                                     |
| NI             | Net Income (Loss)                                              | Net Income plus Total Extraordinary Items Net of Taxes                                                                                                                                                 |
| OANCF          | Operating Activities – Net Cash Flow (Statement of Cash Flows) | Operating Activities – Net Cash Flow (Statement of Cash Flows)                                                                                                                                         |
| OIADP          | Operating Income After Depreciation                            | Operating Income Taxes – Total plus Operating Income – Total                                                                                                                                           |
| PI             | Pretax Income                                                  | Current Operating Earnings Before Income Tax plus Net Pretax Profit or Loss on Securities Sold or Redeemed                                                                                             |
| PSTK           | Preferred Stock – Carrying Value                               | Preferred Stock – Par Value                                                                                                                                                                            |
| RECT           | Receivables – Total                                            | Receivables (Net) plus Accrued Revenues Unbilled                                                                                                                                                       |
| SALE           | Sales (Net)                                                    | Operating Revenues – Total (Income Statement)                                                                                                                                                          |
| TXT            | Income Taxes – Total                                           | Operating Income Taxes – Total plus Nonoperating Income Taxes (Net)                                                                                                                                    |
| XSGA           | Selling, General, and Administrative Expense                   | Salaries and Wages of Officers and Employees plus Pension and Employee Benefits plus Occupancy Expense of Bank Premises – Net plus Furniture and Equipment: Depreciation, Rental Cost, Servicing, Etc. |

**source: Chapter 4 Company Data and Table 1. Annual Xpressfeed Items Compared to Compustat Utility and Compustat Bank Items, Understanding the Data, Standard & Poor’s, 2011*
